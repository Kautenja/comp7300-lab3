
````python
from pandas import DataFrame


def matrix(dimension: int):
    _matrix = []
    x = 0
    for i in range(dimension):
        # add a row to the matrix
        _matrix.append([])
        for j in range(dimension):
            if i >= j:
                _matrix[-1].append(x)
                x += 1
            else:
                _matrix[-1].append(1)
    return DataFrame(_matrix)
````

```python
from pandas import DataFrame
N = 16
```

# Rowwise

```python
def rowwise_nodes(n):
    return n * (n + 1) / 2
```

```python
rowwise_nodes(5 - 1)
```

```python
def rowwise(i, j):
    if i < j:
        return 1
    return rowwise_nodes(i) + j
```

```python
rowwise(1, 0)
```

```python
DataFrame([[rowwise(x, y) for y in range(N)] for x in range(N)])
```

# Columnwise

4x4 - 9

5x5 - 11

6x6 - 13

```python
def t(n):
    return 2 * (n - 4) + 9
```

```python
def columnwise_base(i, n=N):
    return (1 / 2) * (-(i + 1)**2 + t(n)*(i + 1) - 2)
```

```python
def columnwise(i, j, n=N):
    if i > j:
        return 1
    return columnwise_base(i, n) - (n - j - 1)
```

```python
DataFrame([[columnwise(x, y) for x in range(N)] for y in range(N)])
```
